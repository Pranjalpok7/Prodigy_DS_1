```python
netflix_df = pd.DataFrame(netflix_dataset) 
```

```python
netflix_df
```

```python
# Filter the DataFrame to keep only rows where "type" is "Movie"
movies_df = netflix_df[netflix_df["type"] == "Movie"]
```

```python
movies_df
```

```python
import pandas as pd

# Assuming your DataFrame is named "movies_df"

# Extract the numeric part of the duration using regular expressions
movies_df["duration"] = movies_df["duration"].str.extract('(\d+)')

# Convert the extracted strings to numeric values
movies_df["duration"] = pd.to_numeric(movies_df["duration"])
```

```python
# Group data by rating and calculate average duration
grouped_data = movies_df.groupby("rating")["duration"].mean().reset_index()

# Create barplot
plt.bar(grouped_data["rating"], grouped_data["duration"])
plt.xlabel("Rating")
plt.ylabel("Average Duration (minutes)")
plt.title("Average Movie Duration by Rating")
plt.show()
```

```python
# Create scatterplot
plt.scatter(movies_df["rating"], movies_df["duration"])
plt.xlabel("Rating")
plt.ylabel("Duration (minutes)")
plt.title("Rating vs Duration Scatterplot")
plt.show()
```
