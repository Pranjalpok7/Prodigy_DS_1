## The below comparisons shows that Movies are more than double the TV shows in Netflix

```python
type_counts = netflix_df["type"].value_counts()

# Create a bar chart
plt.bar(type_counts.index, type_counts.values)
plt.xlabel("Type")
plt.ylabel("Count")
plt.title("Number of Movies and TV Shows in Netflix Dataset")
plt.show()
```
